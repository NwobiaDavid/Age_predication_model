importing the needed packages

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import joblib
```

loading the dataset

```python
data = pd.read_csv("bmi.csv")
```

extracting the needed data

```python
X = data[['Age']].values
y_weight = data['Weight'].values
y_height = data['Height'].values

print(f"age: {X}, weight: {y_weight}, height: {y_height}")
```

spiting the data into training and teaching set

```python
X_train, X_test, y_weight_train, y_weight_test = train_test_split(X, y_weight, test_size=0.2, random_state=42)

_, _, y_height_train, y_height_test = train_test_split(X, y_height, test_size=0.2, random_state=42)
```

creating and training the model(linear regression model) using only the weight data

```python
model_weight = LinearRegression()
model_weight.fit(X_train, y_weight_train)
```

making predictions on the test set

```python
y_weight_pred = model_weight.predict(X_test)
```

evaluating the weight model to check for errors

```python
mse_weight = mean_squared_error(y_weight_test, y_weight_pred)
print("Mean Squared Error for Weight:", mse_weight)
```

visualization of model

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(X_test, y_weight_test, color='blue', label='Actual Weight')
plt.plot(X_test, y_weight_pred, color='red', linewidth=2, label='Predicted Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Weight Prediction')
plt.legend()

plt.tight_layout()
plt.show()
```

creating and training the model(linear regression model) using only the height data

```python
model_height = LinearRegression()
model_height.fit(X_train, y_height_train)
```

making predictions on the test set

```python
y_height_pred = model_height.predict(X_test)
```

evaluating the height model to check for errors

```python
mse_height = mean_squared_error(y_height_test, y_height_pred)
print("Mean Squared Error for Height:", mse_height)
```

visualization of the model

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_height_test, color='green', label='Actual Height')
plt.plot(X_test, y_height_pred, color='orange', linewidth=2, label='Predicted Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Height Prediction')
plt.legend()

plt.tight_layout()
plt.show()
```

allowing user's input

```python
age = float(input("Enter the age: "))
```

predicting weight and height from the user's input

```python
weight_prediction = model_weight.predict([[age]])
height_prediction = model_height.predict([[age]])

print("Predicted Weight:", weight_prediction[0])
print("Predicted Height:", height_prediction[0])
```
